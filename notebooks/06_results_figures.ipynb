# 06 — Results Figures

Generates the bar charts used in the report and demo video from `results/tables/*.csv`.

```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

fig_dir = Path('results/figures'); fig_dir.mkdir(parents=True, exist_ok=True)
```

```python
report = pd.read_csv('results/tables/report_metrics.csv')
ax = report.set_index('config')[['bleu4', 'rouge_l', 'bertscore_f1', 'chexbert_f1']].plot.bar(figsize=(10,5))
ax.set_title('Report Generation — by config')
ax.set_ylabel('score')
plt.tight_layout(); plt.savefig(fig_dir / 'report_metrics.png', dpi=150)
```

```python
qa = pd.read_csv('results/tables/qa_metrics.csv')
ax = qa.set_index('config')[['exact_match', 'token_f1', 'bertscore_f1', 'llm_judge_mean']].plot.bar(figsize=(10,5))
ax.set_title('QA — by config')
ax.set_ylabel('score')
plt.tight_layout(); plt.savefig(fig_dir / 'qa_metrics.png', dpi=150)
```

```python
ret = pd.read_csv('results/tables/retrieval_metrics.csv')
ax = ret.set_index('backend')[['recall_at_1', 'recall_at_5', 'recall_at_10', 'mrr', 'ndcg_at_10']].plot.bar(figsize=(10,5))
ax.set_title('Retrieval — Recall@K, MRR, nDCG')
plt.tight_layout(); plt.savefig(fig_dir / 'retrieval_metrics.png', dpi=150)
```
